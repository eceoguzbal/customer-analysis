```python
products = (
    df[["StockCode", "Description", "UnitPrice"]]
    .drop_duplicates()
    .rename(columns={"StockCode": "product_id", "Description": "description", "UnitPrice": "price"})
)

products.head()
```

```python
orders = (
    df.dropna(subset=["CustomerID"])[["InvoiceNo", "CustomerID", "StockCode", "InvoiceDate", "Quantity"]]
    .rename(columns={
        "InvoiceNo": "order_id",
        "CustomerID": "customer_id",
        "StockCode": "product_id",
        "InvoiceDate": "order_date",
        "Quantity": "quantity"
    })
)

orders.head()
```

```python
conn = sqlite3.connect("company.db")

customers.to_sql("customers", conn, if_exists="replace", index=False)
products.to_sql("products", conn, if_exists="replace", index=False)
orders.to_sql("orders", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("""
SELECT
  SUM(o.quantity * p.price) AS total_revenue
FROM orders o
JOIN products p
  ON o.product_id = p.product_id
""", conn)
```

```python
pd.read_sql("""
SELECT
  p.product_id,
  p.description,
  SUM(o.quantity * p.price) AS revenue
FROM orders o
JOIN products p
  ON o.product_id = p.product_id
GROUP BY p.product_id, p.description
ORDER BY revenue DESC
LIMIT 10
""", conn)
```

```python
pd.read_sql("""
SELECT
  c.country,
  SUM(o.quantity * p.price) AS revenue
FROM orders o
JOIN customers c
  ON o.customer_id = c.customer_id
JOIN products p
  ON o.product_id = p.product_id
GROUP BY c.country
ORDER BY revenue DESC
LIMIT 15
""", conn)
```

```python
monthly = pd.read_sql("""
SELECT
  substr(o.order_date, 1, 7) AS month,
  SUM(o.quantity * p.price) AS revenue
FROM orders o
JOIN products p
  ON o.product_id = p.product_id
GROUP BY month
ORDER BY month
""", conn)

monthly.head()
```

```python
import matplotlib.pyplot as plt

monthly["month"] = monthly["month"].astype(str)

plt.figure()
plt.plot(monthly["month"], monthly["revenue"])
plt.title("Monthly Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
kpi = pd.read_sql("""
SELECT
    COUNT(DISTINCT o.order_id) AS total_orders,
    COUNT(DISTINCT o.customer_id) AS total_customers,
    SUM(o.quantity * p.price) AS total_revenue,
    SUM(o.quantity) AS total_items_sold
FROM orders o
JOIN products p
    ON o.product_id = p.product_id
""", conn)

kpi
```

```python
rfm = pd.read_sql("""
SELECT
    o.customer_id,
    COUNT(DISTINCT o.order_id) AS frequency,
    SUM(o.quantity * p.price) AS monetary,
    MAX(o.order_date) AS last_order_date
FROM orders o
JOIN products p
    ON o.product_id = p.product_id
GROUP BY o.customer_id
""", conn)

rfm.head()
```

```python
rfm["last_order_date"] = pd.to_datetime(rfm["last_order_date"])

reference_date = rfm["last_order_date"].max()

rfm["recency"] = (reference_date - rfm["last_order_date"]).dt.days

rfm.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features = rfm[["frequency", "monetary", "recency"]]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

kmeans = KMeans(n_clusters=4, random_state=42)

rfm["cluster"] = kmeans.fit_predict(scaled_features)

rfm.head()
```

```python
rfm.groupby("cluster")[["frequency", "monetary", "recency"]].mean()
```
